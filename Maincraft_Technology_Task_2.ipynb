```python
#importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df=sns.load_dataset('titanic')
titanic_df
```

```python
titanic_df.dtypes
```

```python
#handling missing values
titanic_df.isnull().sum()
```

```python
#age
titanic_df['age'].fillna(titanic_df['age'].mean(),inplace=True)
```

```python
#deck
titanic_df['deck']=titanic_df['deck'].astype('object')
titanic_df['deck'].fillna('No-Infomation',inplace=True)
```

```python
# embarked_town,embarked
titanic_df['embarked'].fillna('No-Infomation',inplace=True)
titanic_df['embark_town'].fillna('No-Infomation',inplace=True)
```

```python
titanic_df.isnull().sum()
```

```python
#Answer questions such as:
#Who survived more: males or females?
#Did passenger class affect survival chances?
#What was the survival rate by age group?
```

```python
#Who survived more: males or females?
titanic_df.groupby(by=['sex','survived'])['survived'].count()
```

```python
titanic_df
```

```python
#Did passenger class affect survival chances?
titanic_df.pivot_table(index='class',columns='survived',values='pclass',aggfunc='count')
```

```python
titanic_df['age']=titanic_df['age'].astype('int64')

# creating a user defined function to create an age group
def age_group(age):
  if age<=18:
    return '0-18'
  elif age>18 and age<=30:
    return '19-30'
  elif age>30 and age<=40:
    return '31-40'
  elif age>40 and age<=50:
    return '41-50'
  elif age>50 and age<=60:
    return '51-60'
  else:
    return '60+'
titanic_df['age_group']=titanic_df['age'].apply(age_group)
titanic_df
```

```python
#What was the survival rate by age group?
np.round(titanic_df.groupby(by=['age_group','survived'])['survived'].count()/titanic_df['age_group'].value_counts()*100,2).unstack()
```

```python
#Visualize using Seaborn / Matplotlib:
#o Bar chart of survival by gender.
#o Bar chart of survival by class.
#o Histogram of passenger ages.
```

```python
#o Bar chart of survival by gender.
titanic_df.groupby(by=['sex','survived'])['survived'].count().unstack().plot(kind='bar')
```

```python
#o Bar chart of survival by class.
plt.figure(figsize=(10,7))
sns.barplot(data=titanic_df,x=titanic_df['class'],y=titanic_df['survived'],hue='sex')
plt.title('Survival on The Basis of Passenger Class')
plt.xlabel('Class')
plt.ylabel('No Of Passangers')
plt.xticks(rotation=90)
plt.show()
```

```python
#o Histogram of passenger ages.
plt.figure(figsize=(10,7))
sns.histplot(data=titanic_df,x=titanic_df['age'],kde=True)
```

```python
#Findings:
#1. The missing values in columns like age, embarked, embark_town has been handled using fillna where the missing values in age has been filled with the mean value of the age
#   and for embarked and embark_town it is filled with No-Information.
#2. The chances of survival are higher in females as compare to males.
#3. The first class and second class survival chances are higher as comapre to third class.
#4. The older you are the survival chances are less and the younger you are the survival chances are higher.
```
